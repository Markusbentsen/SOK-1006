### Analyse av en merverdiavgift ("ad valorem tax")

Her skal vi se på virkningen av å ilegge en vare en skatt i henhold til dens verdi (merverdiavgift). Betrakt fremstillingen av markedet for oppdrettslaks i læreboka s. 101-104. Etterspørsel er gitt ved


\begin{equation}
   x^{D}= 500 -3.2p_k\tag{1}
\end{equation}

og tilbud:


\begin{equation}
   x^{S}= -100 + 4.3p_s\tag{2}
\end{equation}

Kjøpere betaler $p_k$ og selgere får $p_s$. Prisen som selgere får er det som konsumenter betaler minus mva $tp_k$, noe som betyr at $p_s=(1-t)p_k$. Her er $1>t\geq 0$ avgiftssatsen. Vi kan skrive (2) også som en funksjon på prisen som konsumenten betaler:

\begin{equation}
   x^{S}= -100 + 4.3 p_k(1-t)\tag{3}
\end{equation}



```python
# importer pakkene
import numpy as np
from matplotlib import pyplot as plt
import sympy as sp
```

```python
# Etterspørsel i likning (1)

def x_d(p):
    return 500 - 3.2*p

#Tilbud i likning (3)

def x_s(p,t):
    return -100+4.3*p*(1-t)
```

```python
# vi lager en basisfigur

p_num=np.linspace(0.1,100,100)

def create_ax():
                fig, ax = plt.subplots()
                ax.set_ylabel('Pris per kilo', loc='top')
                ax.set_xlabel('Tonn per uke', loc='right')
                ax.set(xlim=(75,280))
                ax.set(ylim=(40,100))
                ax.spines['top'].set_color('none')
                ax.spines['right'].set_color('none')
                
                return fig, ax

fig, ax = create_ax()

# plott funksjonene

ax.plot(x_d(p_num),p_num, label='Etterspørsel')
ax.plot(x_s(p_num, 0),p_num, label='Tilbud')


ax.legend(bbox_to_anchor=(1.5,0.6));
```

```python
Merverdiavgift (mva) øker kostnadene for produkter og tjenester, noe som fører til en høyere pris for forbrukeren. 
Når prisen øker, vil etterspørselen etter produktet sannsynligvis gå ned, noe som resulterer i en brattere tilbudsfunksjon. 
På grunn av den økte prisen vil produsentene måtte selge færre enheter for å oppnå samme fortjeneste, noe som gjør at 
tilbudsfunksjonen blir brattere.
```

```python
#Oppgave 1, tegn tilbudsfunksjon

fig, ax = create_ax()

ax.plot(x_d(p_num),p_num, label='Etterspørsel')
ax.plot(x_s(p_num, 0),p_num, label='Tilbud')
ax.plot(x_s(p_num, +0.25),p_num, label='Tilbud med 25% mva')

ax.legend(bbox_to_anchor=(1.5,0.6));
```

```python
#Oppgave 2, likevektspris og kvantum med og uten mva
x,p=sp.symbols('x p', positive=True, real=True)
eq_LS = sp.Eq(500 - 3.2*p, -100 + 4.3*p) 
eq_LNS = sp.Eq(500 - 3.2*p, -100 + 4.3*0.75*p) 
likevekt_s = sp.solve(eq_LS, p) 
likevekt_ns = sp.solve(eq_LNS, p)

#Likevekt_s er 80
#Likevekt_ns er 93.38
```

```python
likevekt_s = round(likevekt_s(0,0),2)
likevekt_ns = round(likevekt_ns(4,0),2)
```

```python
#Oppgave 2, likevektspris og kvantum med og uten mva 

fig, ax = create_ax()

#lager graf
ax.plot(x_d(p_num),p_num, label='Etterspørsel') 
ax.plot(x_s(p_num, 0),p_num, label='Tilbud')
ax.plot(x_s(p_num, 0.25),p_num, label='Tilbud med 25% mva')
ax.plot ([0, x_s(likevekt_s[0], 0)], [likevekt_s[0], likevekt_s[0]], color="black")
ax.plot ([x_s(likevekt_s[0], 0), x_s(likevekt_s[0], 0)], [0, likevekt_s[0]], color="black")
ax.plot ([0, x_s(likevekt_ns[0], 0.25)], [likevekt_ns[0], likevekt_ns[0]], color="black") 
ax.plot ([x_s(likevekt_ns[0], 0.25), x_s(likevekt_ns[0], 0.25)], [0, likevekt_ns[0]], color="black")

ax.legend(bbox_to_anchor=(1.5,0.6)); 


#Likevekt_s er 80 (Tilbud)
#Likevekt_ns er 93.38 (Tilbud med 25% mva)
```
